```python
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as Soup
from bs4 import Tag
import re

images = [cv2.imread(file) for file in glob.glob("/content/drive/MyDrive/screenshots/*.png,*.jpg")]

for file in glob.glob("/content/drive/MyDrive/screenshots/*.png"):
  hocr = pytesseract.image_to_pdf_or_hocr(file, extension='hocr')

  data = hocr.decode("utf-8")

  soup = Soup(data, 'html.parser')
  font_spans = [ data for data in soup.select('span') if 'x_size' in str(data) ]
  output = []
  for i in font_spans:
    tup = ()
    fonts_size = re.search(r'(x_size)( .*?)[ 0-9.]+',str(i)).group()
    fonts_size = re.sub(r"x_size ", '' , fonts_size)

    tup = (str(i.text).strip(), fonts_size.strip())
    output.append(tup)
  print(file)     
  print(output)
```

```python
import pytesseract
from PIL import Image
import cv2
img = cv2.imread('/content/drive/MyDrive/screenshots/cityhomestead.png')

print(pytesseract.image_to_string(img))
hImg, wImg, _ = img.shape

boxes = pytesseract.image_to_boxes(img)

for b in boxes.splitlines():
  b = b.split(' ')
  print(b)
x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
cv2.rectangle(img, (x, hImg - y), (w, hImg - h), (50, 50, 255), 1)
cv2.putText(img, b[0], (x, hImg - y + 13), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (50, 205, 50), 8)
```
